## <font color='darkblue'>Generadores de números pseudoaleatorios</font>
Este apartado contiene varios métodos para generar números pseudoaleatorios, los cuales se presentan a continuación.

Para cada generador, se tiene una breve explicación, definición y ejemplos de uso.

### <font color='blue'>Generador congruencial aditivo</font>

Dada una semilla $X_0 \in Z^{+}$, dos números $m \in N$ y $a \in Z^{+}$ con $a < m$, se define para $i = 1, 2, 3, \ldots $:

$$X_i = (X_{i−1} + a) \text{ mod } m$$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{X_i}{m}.$

###### Uso

- `Congruencial_aditivo(x0, b, m).random()` regresa un número pseudoaleatorio
- `Congruencial_aditivo(x0, b, m).randoms()`  regresa todos los números pseudoaleatorios posibles sin números repetidos

```python
generador = Congruencial_aditivo(15, 96, 74)
```

```python
generador.random()
```

```python
generador.randoms()[5:10]
```

### <font color='blue'>Generador congruencial multiplicativo</font>


Dada una semilla $X_0 \in Z^{+}$, dos números $m \in N$ y $a \in Z^{+}$ con $a < m$, se define para $i = 1, 2, 3, \ldots $:

$$X_i = (X_{i−1} \cdot a) \text{ mod } m$$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{X_i}{m}.$

###### Uso

- `Congruencial_multiplicativo(x0, b, m).random()` regresa un número pseudoaleatorio
- `Congruencial_multiplicativo(x0, b, m).randoms()`  regresa todos los números pseudoaleatorios posibles sin números repetidos

```python
generador = Congruencial_multiplicativo(89, 148, 985)
```

```python
generador.random()
```

```python
generador.randoms()[5:10]
```

### <font color='blue'>Generador congruencial polinomial</font>

Dada una semilla $X_0 \in Z^{+}, k + 1$ multiplicadores $b_0, b_1, \ldots , b_k (k \geq 2)$ donde $b_0,\ldots , b_k \in Z^{+}$ con $X_0, b_0, . . . , b_k < m$, definimos para $i = 1, 2, 3,\ldots$

$$X_i = (b_0 + b_1 X_{i−1} + b_2 X^{2}_{i−1} + \cdots + b_k X^{k}_{i−1}) \text{ mod } m$$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{X_i}{m}$.


###### Uso

- `Congruencial_polinomial(x0, b_s, m).random()` regresa un número pseudoaleatorio
- `Congruencial_polinomial(x0, b_s, m).randoms()`  regresa todos los números pseudoaleatorios posibles sin números repetidos

```python
generador = Congruencial_polinomial(1, [451, 163], 175)
```

```python
generador.random()
```

```python
generador.randoms()[0:5]
```

### <font color='blue'>Generador congruencial multiplicativo múltiple</font>

Dadas $k$ semillas $X_{01}, \ldots, X_{0k}, k$ multiplicadores $b_1,\ldots , b_k$ y una $m$ con
$X_{0j}, b_{j}, m \in Z^{+}$ para $j = 1, \ldots, k$ y $k \geq 2$, definimos para $i = −k + 1, −k + 2, −k + 3,\ldots$

$$ X_{i+k} = \left(b_1 X_{i+k−1} + b_2 X_{i+k−2} + \cdots + b_k X_i\right)  m$$

donde $X_0 = X_{01}, X_{-1} = X_{02}, \ldots , X_{−k+1} = X_{0k}$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{X_i}{m}.$

###### Uso

- `Congruencial_multiplicativo_multiple(x0_s, b_s, m).random()` regresa un número pseudoaleatorio
- `Congruencial_multiplicativo_multiple(x0_s, b_s, m).randoms(n)`  regresa $n$ números pseudoaleatorios sin repeticiones si es posible, en caso contrario, devuelve los posibles sin repeticiones.

```python
generador = Congruencial_multiplicativo_multiple([659, 155, 728], [587, 121, 947], 777)
```

```python
generador.random()
```

```python
generador.randoms(5)
```

### <font color='blue'>Generador Congruencial Lineal Combinado</font>

Dadas dos semillas $X_0, Y_0 \in Z^{+}$, dos multiplicadores $b_1$ y $b_2$ y dos módulos $m_1$ y $m_2$ con $b_1, b_2, m_1, m_2 \in Z^{+}$, definimos para $i = 1, 2, 3, \ldots$

$$Xi = b_1 X_{i−1} \text{ mod } m_1$$
$$Yi = b_2 X_{i−1} \text{ mod } m_2$$
$$Zi = (X_i − Y_i) \text{ mod } m_1$$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{Z_i}{m_1}$.

###### Uso

- `Congruencial_combinado(x0, y0, b1, b2, m1, m2).random()` regresa un número pseudoaleatorio
- `Congruencial_combinado(x0, y0, b1, b2, m1, m2).randoms()`  regresa todos los números pseudoaleatorios posibles sin números repetidos

```python
generador = Congruencial_combinado(1, 7, 436, 831, 794, 197)
```

### <font color='blue'>Generador Congruencial Combinado Múltiple</font>

Dadas $2k$ semillas $X_{01}, \ldots , X_{0k}, Y_{01}, \ldots , Y_{0k} \in Z^{+}, 2k$ multiplicadores $b_1,\ldots , b_k,
c_1, \ldots, c_k$ y $m_1, m_2$, definimos para $i = −k + 1, −k + 2, −k + 3,\ldots$

$$X_{i+k} = (b_1 X_{i+k−1} + b_2 X_{i+k−2} + \ldots + b_k X_i) \text{ mod } m_1$$
$$Y_{i+k} = (c_1 X_{i+k−1} + c_2 X_{i+k−2} + \ldots + c_k X_i) \text{ mod } m_2$$
$$Z_{i+k} = (X_{i + k} − Y_{i+k}) \text{ mod } m_1$$

por lo que el i-ésimo número pseudoaleatorio queda definido como $u_i = \frac{Z_i}{m_1}$.

###### Uso

- `Congruencial_combinado_multiple(x0_s, y0_s, b1_s, b2_s, m1, m2).randoms()`  regresa todos los números pseudoaleatorios posibles sin números repetidos

La función de masa de probabilidad de la distribución binomial se define como:

- $ p $ es la probabilidad de éxito en un solo ensayo.
- $ n $ es el número total de ensayos.


###### Uso

- `Numgen.binomial(n, p, N)` regresa $N$ números de la distribución

### <font color='blue'>Poisson</font>

La función de distribución de Poisson describe la probabilidad de que ocurran exactamente $k$ eventos en un intervalo fijo de tiempo o espacio, dado que los eventos ocurren con una tasa promedio constante $ \lambda $. La fórmula de la función de distribución de Poisson es:

$$
P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}
$$

Donde:
- $ e $ representa al número de Euler
- $ \lambda $ es la tasa promedio de eventos por unidad de tiempo o espacio.
- $ k $ es el número de eventos.


###### Uso

###  <font color='blue'>T-Student</font>

Sea $X$ una variable aleatoria con distribución $\chi^2_k$, y sea $Z$ una variable aleatoria con distribución $N(0, 1)$, entonces, la variable aleatoria $T = \frac{Z}{\sqrt{\frac{\chi^2_k}{k}}}$ tiene distribución T-student, cuya función de densidad es la siguiente: $
f(x; k) = \frac{\Gamma((k+1)/2)}{\sqrt{\pi k} \Gamma(k/2)} \left(1 + \frac{x^2}{k}\right)^{-(k+1)/2}
$

###### Uso

### <font color='blue'>Modelo de Black and Scholes</font>

###### Uso

- `B_S(S_0, K, t_0, T, r, sigma_estimation, data).call()` regresa el precio teórico de una opción call


- `B_S(S_0, K, t_0, T, r, sigma_estimation, data).put()` regresa el precio teórico de una opción put


- `B_S(S_0, K, t_0, T, r, sigma_estimation, data).plot_profit()` muestra el gráfico del payoff de las opciones


- `B_S(S_0, K, t_0, T, r, sigma_estimation, data).plot_prices()` muestra el gráfico de los precios de ambas opciones call y put, variando el precio


- `B_S(S_0, K, t_0, T, r, sigma_estimation, data).plot_time()` muestra el gráfico de los precios de ambas opciones call y put, variando el tiempo de expiración

donde
- $S_0$ representa el precio de la acción al momento 0
- $K$ representa el precio del ejercicio
- $T$ representa la fecha de expiración del contrato de la opción
- $r$ representa la tasa libre de riesgo 
- $sigma_estimation$ representa el método para estimar a la volatilidad. Se incluye estimación histórica o por su función de densidad. ('hist', 'density')
- $data$ representa un dataframe, en donde se tiene una columna con los log rendimientos de la acción, en donde dicha columna se deberá llamar: `Log Returns `
